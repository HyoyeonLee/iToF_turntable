# defs

1) Chart
- h : height of chart
- w : width of chart

2) FOV
- Ver : vertical FOV
- Hor : horizontal FOV

3) FOV angle
- $\theta_V$ : vertical FOV in angle
- $\theta_H$ = 100$^\circ$ : horizontal FOV in angle

4) distance
- d=500 : distance from the lens to the Center of the chart
- $\Delta d = \frac{w}{2}sin\alpha$ : distance between the center and the edge of the chart (horizontal)
- D = d+$\Delta d$

5) angle
- $\alpha$ =30$^\circ$: 




```python
import numpy as np
from matplotlib import pyplot as plt
def rad(angle):
    return angle*np.pi/180.0
w = np.linspace(1000,3000,500)
d = 400
alpha = 30
dd = abs(w*0.5*np.sin(rad(alpha)))
D = d+dd

theta_hor =100
Hor = 2*D*np.tan(rad(theta_hor/2.0))
fig = plt.figure(figsize=(10,10))
plt.minorticks_on()
plt.grid(b=True, which='major', color='lightgrey', linestyle='--')
plt.grid(b=True, which='minor', color='lightgrey', linestyle='--')
plt.scatter(w,Hor,label='FOV')
plt.plot(w,w,'r',linestyle='--',label = 'Chart Width')
plt.xlabel('Chart Width [mm]',fontsize=20,weight='bold')
plt.ylabel('FOV[mm]',fontsize=20,weight='bold')
plt.scatter(w,500+w*0.5/2,label='Distance')
plt.axvline(x =1610 ,linestyle='--', color = 'lightgrey')
#plt.axvline(x =w[432] ,linestyle='--', color = 'g') #theta_H=98.2
#plt.axhline(y =Hor[432] ,linestyle='--', color = 'g')

#plt.axvline(x =w[487] ,linestyle='--', color = 'g')#theta_H = 100, d=500
#plt.axhline(y =Hor[487] ,linestyle='--', color = 'g')
plt.axvline(x =w[340] ,linestyle='--', color = 'g')#theta_H = 100, d=400
plt.axhline(y =Hor[340] ,linestyle='--', color = 'g')

#plt.axvline(x =w[152] ,linestyle='--', color = 'g')#theta_H = 83.2
#plt.axhline(y =Hor[152] ,linestyle='--', color = 'g')

plt.legend()
```

```python
for i in range(len(w)):
    chart = w[i]
    fov = Hor[i]
    if w[i]>=1608 and w[i]<1612:
        print(i,w[i])
    if chart>=fov:
        print(i)
        break
w0 = chart
h0 = w0*3/4
d0 = np.sqrt(w0**2+h0**2)
print(f"chart = {chart}, fov= {fov}")
print(f"chart : {np.round(w0)} x {np.round(h0)}, diagonal = {np.round(d0)}")
```
